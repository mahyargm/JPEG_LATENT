```python
from PIL import Image
import os

# Read image height and width
image_path = '/workspace/jpeg/results/test/rec/JAI_00002_TE_2144x1424_8bit_sRGB_012.png'
with Image.open(image_path) as img:
    width, height = img.size

print(f"Image width: {width}, height: {height}")

# Read bit file size in bits
bit_file_path = '/workspace/jpeg/results/test/bit/JAI_00002_TE_012.bits'
bit_file_size_bytes = os.path.getsize(bit_file_path)
bit_file_size_bits = bit_file_size_bytes * 8

print(f"Bit file size: {bit_file_size_bits} bits")

print("Actual bpp of the image:", bit_file_size_bits / (width * height))
```
